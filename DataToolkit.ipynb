# **NumPy, Pandas & Visualization – Q&A**
##1. What is NumPy, and why is it widely used in Python?
*  NumPy is a powerful library for numerical computations in Python. It provides efficient array operations, supports multi-dimensional arrays, and is much faster than Python lists due to its C-based backend and vectorized operations.

##2. How does broadcasting work in NumPy?
* Broadcasting enables arithmetic operations on arrays of different shapes. NumPy automatically expands the dimensions of the smaller array to match the larger one without copying data, allowing element-wise operations.

##3. What is a Pandas DataFrame?
* A DataFrame is a 2-dimensional labeled data structure in Pandas, similar to a table or spreadsheet, with rows and columns. It supports heterogeneous data types and offers powerful data manipulation tools.

##4. Explain the use of the groupby() method in Pandas.
* The groupby() method splits data into groups based on some criteria, applies a function to each group (e.g., sum, mean), and then combines the results. It’s useful for aggregation and summarization.

##5. Why is Seaborn preferred for statistical visualizations?
* Seaborn is built on top of Matplotlib and offers a high-level interface for drawing attractive and informative statistical graphics. It simplifies complex plots like boxplots, violin plots, and heatmaps.

##6. What are the differences between NumPy arrays and Python lists?
* NumPy arrays are faster, consume less memory, and support element-wise operations. Python lists are more flexible but slower and inefficient for large-scale numerical operations.

##7. What is a heatmap, and when should it be used?
* A heatmap is a data visualization that uses color to represent values in a matrix. It’s ideal for showing correlation matrices, feature importance, or missing data patterns.

##8. What does the term “vectorized operation” mean in NumPy?
* Vectorized operations are operations performed on entire arrays without using loops. They are faster and more efficient because they utilize low-level optimizations.

##9. How does Matplotlib differ from Plotly?
* Matplotlib is a static plotting library ideal for traditional graphs, while Plotly offers interactive, web-based plots with features like zooming and tooltips. Plotly is better for dashboards and presentations.

##10. What is the significance of hierarchical indexing in Pandas?
* Hierarchical indexing allows multiple index levels (MultiIndex) on rows and/or columns. It supports complex data structures and enables easier data slicing and aggregation.

##11. What is the role of Seaborn’s pairplot() function?
* pairplot() creates scatterplots for all pairwise combinations of numerical features in a dataset, helping to visualize relationships and detect patterns or correlations.

##12. What is the purpose of the describe() function in Pandas?
* describe() generates summary statistics for numerical columns, such as count, mean, std, min, and percentiles. It's useful for quick exploratory data analysis.

##13. Why is handling missing data important in Pandas?
* Handling missing data ensures accuracy in analysis. Pandas provides functions like fillna() and dropna() to deal with missing values to prevent biased or invalid results.

##14. What are the benefits of using Plotly for data visualization?
* Plotly offers interactive, zoomable, and customizable plots suitable for dashboards and web apps. It supports 3D plots, maps, and real-time updates.

##15. How does NumPy handle multidimensional arrays?
* NumPy supports n-dimensional arrays (ndarray) and provides functions for reshaping, slicing, broadcasting, and applying operations across any axis.

##16. What is the role of Bokeh in data visualization?
* Bokeh is a Python interactive visualization library that creates rich, web-based plots and dashboards. It is ideal for streaming and large-scale interactive data apps.

##17. Explain the difference between apply() and map() in Pandas.
* map() is used for element-wise operations on a Series, while apply() can apply a function to each row or column of a DataFrame or to a Series. apply() is more versatile.

##18. What are some advanced features of NumPy?
* Advanced features include broadcasting, masked arrays, memory-mapped files, linear algebra functions, random sampling, and Fourier transforms.

##19. How does Pandas simplify time series analysis?
* Pandas provides time-aware data structures, date-range generation, resampling, frequency conversion, and powerful date-time indexing, making time series analysis efficient.

##20. What is the role of a pivot table in Pandas?
* A pivot table reshapes and summarizes data. It allows grouping, aggregation, and reorganization of data, similar to Excel pivot tables.

##21. Why is NumPy’s array slicing faster than Python’s list slicing?
* NumPy slicing returns views (not copies) of data and is implemented in C, making it significantly faster and memory-efficient compared to Python lists, which return new lists.

##22. What are some common use cases for Seaborn?
* Common use cases include visualizing distributions (e.g., histograms, KDEs), relationships (scatter, line plots), categorical data (bar, box, violin plots), and heatmaps for correlations.

```python
# 1. How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print(row_sum)
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd

df = pd.DataFrame({'Scores': [85, 90, 78, 92]})
mean_score = df['Scores'].mean()
print(mean_score)
```

```python
# 3. Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [5, 6, 7, 8]
})
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# 5. Generate a bar plot using Plotly.
import plotly.express as px
import pandas as pd

data = {'Fruit': ['Apple', 'Banana', 'Cherry'], 'Quantity': [10, 15, 7]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Fruit', y='Quantity', title='Fruit Quantities')
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
df = pd.DataFrame({'Price': [100, 200, 300]})
df['Discounted_Price'] = df['Price'] * 0.9
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 5, 10, 17]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title('Multiple Lines')
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd

df = pd.DataFrame({'Age': [22, 30, 18, 25]})
filtered = df[df['Age'] > 21]
print(filtered)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import numpy as np

data = np.random.randn(1000)
sns.histplot(data, bins=30, kde=True)
```

```python
# 11. Perform matrix multiplication using NumPy.
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = np.dot(A, B)
print(result)
```

```python
# 12. Use Pandas to load a CSV file and display its first 5 rows.

import pandas as pd

# Creating a sample DataFrame and saving it as a CSV file
sample_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 22, 35, 28],
    'City': ['New York', 'Paris', 'London', 'Berlin', 'Tokyo']
}
df_sample = pd.DataFrame(sample_data)
df_sample.to_csv('sample_data.csv', index=False)

# Now loading that CSV and displaying the first 5 rows
df_loaded = pd.read_csv('sample_data.csv')
print(df_loaded.head())
```

```python
# 13. Create a 3D scatter plot using Plotly.
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 11, 12, 13],
    'z': [5, 6, 7, 8]
})
fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
